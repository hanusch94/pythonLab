# Ciklusok

```python
# A ciklusokcélja az ismétlődő feladat szé

print(0)
print(1)
print(2)
print(3)
```

## While
- A while addig fut újra, amíg a feltétele teljesül
- Ha másképp szeretnénk mondani, akkor egy ismétlődő if-ről van szó

```python
a = 0
while(a<3):
    print(a)
    a+=1
```

## For
- Ha egy listán szeretnénk végigmenni, akkor a for ciklus tud megoldást adni számunkra

```python
# Formátum: for [elem] in [lista]:
a = ["a", "b", "c", "d"]
for elem in a:
    print(elem)
```

## For i in range()

```python
# A range() egy iterable (iterálható listát) hoz létre
# Ennek a tényleges belseje végigiterálva (végig lépkedve) oldható meg
print(range(10))
```

```python
# Gyakorlati használatban
for lm in range(5):
    print(lm, end=" ")
```

```python
# A lista kezdőértékét tudjuk változtatni
for lm in range(10,15):
    print(lm, end=" ")
```

```python
# Ha a léptetésen szeretnénk változtatni, akkor a 3. paraméterrel lehet ezt megoldani
for lm in range(10,20, 2):
    print(lm, end=" ")
```

## Enumerate

```python
# Az enumerate arra való, hogy végigmenjünk egy listán úgy, hogy az aktuális elem indexét is tudjuk
# Akkor hasznos, ha valamit csinálni kell az index-szel és az elemmel is

a = ["a", "b", "c", "d"]
for i, lm in enumerate(a):
    print("Az {}. elem a '{}'".format(i+1, lm))
```

```python
# Arra is lehet használni, hogy két összetartozó listán végigmenjünk

a = ["a", "b", "c", "d"]
b = [1, 2, 3, 4]
for i, lm in enumerate(a):
    print(b[i], lm)
```
